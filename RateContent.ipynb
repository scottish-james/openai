# Rate our content 
This script is to demo how we can rate answers to identify articles to re-write. The rating will need to be validated on a sample to see if we agree with ChatGPT and if so scale solution if not try to refine prompt. Scope do this 5 times if sample we still disagree with sample ratings after 5 attempts a decision should be made if a manual review would be faster.

```python
#Import libraries required to run code
import pandas as pd
from openai import OpenAI
import openai

# Read API key from text file
with open('api_key.txt', 'r') as file:
    api_key_1 = file.read().strip()
    
# Initialize the OpenAI client
openai.api_key = api_key_1

#Connect to dummy data
df = pd.read_csv('FAQ_Table.csv')
df.head()

#Fuction to rate answer - refine prompt in message.content
def rate_answer(response):
    """
    Rate the helpfulness of a given response.

    This function takes a response as input and uses the OpenAI GPT-3.5-turbo model
    to rate its helpfulness. The model evaluates the response and returns a numerical
    value between 1 and 5, where 1 indicates the least helpful and 5 indicates the most helpful.

    Parameters
    ----------
    response : str
        The response to be rated for helpfulness.

    Returns
    -------
    str
        A numerical value (as a string) between 1 and 5 representing the helpfulness of the response.

    Examples
    --------
    >>> response = "The current interest rate for savings accounts is 0.5%."
    >>> rate_answer(response)
    '4'

    Notes
    -----
    The function interacts with the OpenAI API to generate the rating. Make sure you have
    configured the OpenAI API key and the `openai` library is installed and imported properly.

    """
    
    prompt = response
    
    completion = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": """
            You are a helpful assistant that 
            reviews how helpful the answers are explained. 
            Only give the numerical value between 1 - 5
            """},
            {"role": "user", "content": prompt}
        ]
    )
    #print(completion)
    rating = completion.choices[0].message.content
    #print(rating)
    return rating

#Loop though column of choice to rate
df['answer_rating'] = df['Answer'].apply(lambda x: rate_answer(x))

#Save rating table
df.to_csv('FAQ_Table_Rating.csv', index=False)

#View rating table
df.head()

```
